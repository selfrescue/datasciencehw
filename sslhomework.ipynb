```python
# Node 클래스 정의
class Node:
	def __init__(self, data):
		self.data = data
		self.next = None

# LinkedList 클래스 정의

class LinkedList:

	# 초기화 메소드
	def __init__(self):
		dummy = Node("dummy")
		self.head = dummy
		self.tail = dummy

		self.current = None
		self.before = None

		self.num_of_data = 0

	# append 메소드 (insert - 맨 뒤에 노드 추가, tail과 node의 next, 데이터 개수 변경)
	def append(self, data):
		new_node = Node(data) 
		self.tail.next = new_node
		self.tail = new_node

		self.num_of_data += 1
        
	# delete 메소드 (delete - current 노드 삭제, 인접 노드의 current, next 변경, 데이터 개수 변경)
	def delete(self):
		pop_data = self.current.data

		if self.current is self.tail:
			self.tail = self.before

			# 중요 : current가 next가 아닌 before로 변경된다.
			self.before.next = self.current.next
			self.current = self.before 
			self.num_of_data -= 1
			return pop_data

	# first 메소드 (search1 - 맨 앞의 노드 검색, before, current 변경)
	def first(self):
		# 데이터가 없는 경우 첫번째 노드도 없기 때문에 None 리턴
		if self.num_of_data == 0: 
			return None

		self.before = self.head
		self.current = self.head.next

		return self.current.data

	# next 메소드 (search2 - current 노드의 다음 노드 검색, 이전에 first 메소드가 한번은 실행되어야 함)
	def next(self):
		if self.current.next == None:
			return None

		self.before = self.current
		self.current = self.current.next

		return self.current.data

	# size 메소드
	def size(self):
		return self.num_of_data 
```

### insert_at 함수 
>조건 1) insert_at(position, new_data) 형식  
조건 2) 리스트의 주어진 위치(position)에 new_data를 삽입한다.    
조건 3) 이 때, 맨 첫 원소의 위치는 1로 정하며, 0 이하의 position 값이 입력되면 error 문을 출력한다.   
조건 4) 만일 position이 현재 리스트의 크기(원소 갯수)보다 크면 맨 마지막에 new_data를 삽입한다. 

- - - 
#### 코드 설명
> 1. 아래 코드는 LinkedList 클래스의 insert_at 메소드로, 지정된 위치인 position에 new_data를 삽입하는 역할을 한다.  <span style="color:red">*(1번 조건 만족)*</span>
2. 아래 코드에서 주석 처리된 1번에서 position 값이 유효한지 확인한다. position이 0 이하의 값일 경우, 조건에 따라 오류 메시지를 출력하고 함수를 종료한다.  <span style="color:red">*(3번 조건 만족)*</span>
3. 2번에서는 position이 현재 리스트의 크기(num_of_data)보다 클 경우, append() 함수를 호출하여 리스트의 맨 마지막에 new_data를 삽입한다.   <span style="color:red">*(4번 조건 만족)*</span>
4. 위 조건이 해당되지 않는 경우 3번부터 실행된다. 3번은 new_data를 저장할 노드를 생성하는 것이다.  
5. 4번에서 현재 위치를 head의 위치로 세팅한다음, position의 횟수만큼 리스트를 순회하며 새로운 노드를 삽입할 위치를 찾는다. 이때, before 노드와 current 노드를 이용하여 위치를 찾게 된다.   <span style="color:red">*(3번 조건 만족)*</span>
6. 5번에서 before.next와 새로운 노드를 연결하고, new_node.next와 current 노드를 연결해 position의 위치에 노드를 삽입한다.   <span style="color:red">*(2번 조건 만족)*</span>
7. new_data가 삽입되었으므로, 6번에서 리스트의 길이(num_of_data)를 1 증가시킨다.   

```python
	def insert_at(self, position, new_data):
    # 1. position이 0 이하의 값일 경우,오류 메시지를 출력하고 함수를 종료한다. 
		if position <= 0:
			print("Error: position은 0 이하가 될 수 없음!")
			return
        
        # 2. position이 현재 리스트의 크기(num_of_data)보다 크면 append() 함수를 호출하여 리스트의 맨 마지막에 new_data를 삽입한다. 
		if position > self.num_of_data:
			self.append(new_data)
			return
        
        # 3. new_data를 저장할 새로운 노드를 생성한다.
		new_node = Node(new_data)

        # 4. 위치를 맨 앞으로 세팅하고, position의 횟수만큼 리스트를 순회하며 새로운 노드를 삽입할 위치를 찾는다.
		self.current = self.head
		self.before = None
		for _ in range(position):
			self.before = self.current
			self.current = self.current.next
            
        # 5. before 노드와 새로운 노드를 연결하고, 새로운 노드와 current 노드를 연결해 position의 위치에 new_node를 삽입한다. 
		self.before.next = new_node
		new_node.next = self.current
        
        # 6. new_data가 삽입되었으므로, 리스트의 길이(num_of_data)를 1 증가시킨다.
		self.num_of_data += 1
        
```

### remove 함수 
>조건 1) remove(key) 형식   
조건 2) 리스트의 원소 가운데 key값과 일치하는 원소를 모두 삭제하고, 리스트를 수정한다.   
조건 3) 처리 결과를 다음 예와 같이 출력한다.    
> * 3번째 원소(key)를 삭제합니다.     
> * 해당하는 원소가 없습니다.



- - - 
#### 코드 설명
> 1. 아래 코드는 LinkedList 클래스의 remove 메소드로, key에 해당하는 값을 가지고 있는 노드를 찾아서 삭제하는 역할을 한다. <span style="color:red">*(1번 조건 만족)*</span>
2. 1번에서 현재 노드를 저장하는 변수인 current를 head, 이전 노드를 저장하는 변수인 prev를 None으로 지정함으로써 위치를 맨 앞으로 세팅한다. 또한, key값을 가진 노드가 있는 지 확인하는 변수인 found와 key값과 같은 data를 가진 노드의 위치를 저장하는 변수인 position, key값을 가진 노드를 확인한 횟수인 found_num을 선언한다. 
3. 2번에서 current가 맨 끝인 None이 될 때까지 반복한다. 만약, current의 data와 key가 일치할 경우 found를 True로 변경하고, found_num에 1을 더한 뒤, 현재 position+found_num-1 번째 원소(key)를 삭제한다는 문장을 출력한다. <span style="color:red">*(3번 조건 만족)*</span>   
  >> found_num - 1 한 값을 더하는 이유는 key 값을 가진 노드가 1개 이상일 경우, 앞의 노드가 삭제 된 이후의 리스트의 인덱스가 출력되기에, 이를 방지하기 위함이다. 
4. 3번에서 current의 data와 key가 일치하는 상태에서 만약 current가 head인 경우,(맨 앞의 값인 경우) head를 다음 노드로 변경하고, current를 head로 지정해 현재 값을 삭제한다.   
  만약 head가 아닌 경우, prev의 next를 current의 next로 변경하고, current를 다음 노드로 바꾸는 방식으로 현재 값을 삭제한다. <span style="color:red">*(2번 조건 만족)*</span>
5. 4번과 같이 current의 data와 key가 일치하지 않는 상태에서는 prev를 current, current를 current.next로 변경해 다음 노드로 이동하며, position을 1 늘린다. 
6. 만약 key 값을 가진 노드가 없을 경우, 해당하는 원소가 없다는 메시지를 출력한다. <span style="color:red">*(3번 조건 만족)*</span>


```python
    def remove(self, key):
        # 1. 현재 위치를 맨 앞으로 세팅한 후, found, position 변수 선언
		current = self.head 
		prev = None
		found = False
		position = 0
		found_num = 0
		
		# 2. 리스트를 순회하면서 key값과 같은 값을 가진 노드를 찾아 해당 원소(key)를 삭제한다고 출력
		while current is not None:
			if current.data == key:
				found = True
				found_num += 1
				print(f'{position + found_num - 1}번째 원소({key})를 삭제합니다.')
                
                # 3. key값과 같은 값을 가진 노드를 삭제                
				if current == self.head:
					self.head = current.next
					current = self.head
				else:
					prev.next = current.next
					current = current.next
					
			else:
				# 4. 다음 노드로 이동
				prev = current
				current = current.next
				position += 1
                
       # 5. key값을 가지는 노드가 없을 경우 출력
		if not found:
			print(f'해당하는 원소가 없습니다.')
```

### traverse_all 함수 
>조건 1) traverse_all() 형식   
조건 2) head부터 tail까지 각 노드를 순차적으로 탐색하며 각 노드의 data를 print한다.  
조건 3) 출력 형식:  head -> (100) -> (72) -> (325) -> null  

- - - 
#### 코드 설명
> 1. 아래 코드는 LinkedList 클래스의 traverse_all 메소드로, head부터 tail까지의 노드를 순차적으로 참색해 노드의 data를 출력하는 역할을 한다. <span style="color:red">*(1번 조건 만족)*</span>
2. current변수에 self.head.next를 할당해 head 다음인 첫번째 노드부터 시작하며, result는 'head -> ({current.data}) -> ({current.data}) -> ... -> null 순으로 출력될 수 있도록, 'head'를 먼저 할당한다. 
3. while루프를 돌며 현재 위치인 current가 None이 아닐 때 까지(맨 끝에 도착할 때 까지) result 문에 '-> (현재 노드의 데이터)' 를 추가하고, current 변수를 current.next로 바꿈으로써 노드 위치를 다음으로 이동한다. <span style="color:red">*(2번 조건 만족)*</span>
4. while문을 빠져나오고 나면 result 문에 tail에 해당하는 '-> null' 을 더한다.
5. 최종적으로 모든 문장이 추가된 result를 출력한다. <span style="color:red">*(3번 조건 만족)*</span>
  >> **head와 tail을 포함하지 않고 별도로 print한 이유는 __init__에서 head와 tail이 dummy로 선언되어 있기 때문이다.** 


```python
# tranverse_all 메소드
def traverse_all(self):  # 모든 노드를 탐색하는 메소드
	current = self.head.next  # head 다음 노드부터 시작
	result = 'head'
	while current is not None:  # 현재 노드가 None이 아닐 때까지 반복
		result += f' -> ({current.data})'  # result에 현재 노드의 data를 추가
		current = current.next  # 다음 노드로 이동
	result += ' -> null'  # 마지막으로 null 추가
	print(result)  # 결과 출력
```
